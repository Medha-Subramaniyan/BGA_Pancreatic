```python
def csv_file_names_to_dataframe(file_names, skiprows):
    data = []
    for idx, item in enumerate(file_names):
        try:
            df = pd.read_csv(item, on_bad_lines='skip', delimiter=',', skiprows=skiprows)  # Adjust delimiter if needed
            data.append(df)
        except pd.errors.ParserError as e:
            print(f"Error reading {item}: {e}")
        except Exception as e:
            print(f"Unexpected error with {item}: {e}")

    if data:
        return pd.concat(data, ignore_index=True)
    else:
        return pd.DataFrame()  # Return empty DataFrame if all files fail
```

To convert a .csv file into a dataframe, do one of the following
1. With a list of file names, run df = csv_file_names_to_dataframe(file_names, X) with file_names being the name of the list holding all of the file names and X being the number of lines to skip from the .csv
2. With a single file name, fun df = pd.read_csv('example.csv')
If that does not work, then the file might have a different delimiter or have certain lines that need to be skipped in order to run properly, in which case you'll need to do something like df = pd.read_csv('example.csv', on_bad_lines='skip', delimiter=',', skiprows=5)

```python
import requests

# URL containing the tables
url = 'https://www.lung.org/research/trends-in-lung-disease/tobacco-trends-brief/overall-smoking-trends'

# Fake a browser visit
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
}

# Make the request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Save the HTML content
    with open("page.html", "w", encoding="utf-8") as f:
        f.write(response.text)

    # Read tables from the saved HTML file
    tables = pd.read_html("page.html")

    # Display the number of tables found
    print(f"Number of tables found: {len(tables)}")

    # Print first few rows of each table
    for i, table in enumerate(tables):
        print(f"\nTable {i + 1}:")
        print(table.head())
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```
